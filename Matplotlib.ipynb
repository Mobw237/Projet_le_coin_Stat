# 1- Importation du package

```python
# Importation des packages

import matplotlib.pyplot as plt 
import numpy as np
```

# 2- Tracons des graphiques simples avec Matplotlib

```python
# Generation des donnees

x = np.linspace(0, 10, 1000)
y = np.exp(x)
```

```python
# Tracons un graphique simple

plt.plot(x, y)
```

# 3- Pesrsonnalisation des graphiques avec Matplotlib

```python
# Changeons la taille d'un graphique avec la largeur et la hauteur de la figure en pouce

plt.figure(figsize = (5, 5))
plt.plot(x, y)
```

```python
# Titre et axes

plt.plot(x, y)
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
```

```python
# Changement de couleur

plt.plot(x, y, c = "green")
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
```

```python
# Changer le type de ligne

plt.plot(x, y, c = "red", ls = "--")
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
```

```python
# Ajoutons une legende 

plt.plot(x, y, c = "red", ls = "-", label = "Exp")
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
```

```python
# Ajoutons une grille sur le graphique

plt.plot(x, y, c = "blue", ls = "--", label = "Exp")
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.grid()
```

# 4- Faire plusieurs graphes

# Faire des graphes sur un meme carre

```python
# Creons des graphes vides

z = 2 * y
y3 = 2 * z
plt.plot(x, y, c = "blue", ls = "--", label = "Exp")
plt.plot(x, z, c = "red", ls = "--", label = "2 * Exp")
plt.plot(x, y3, c = "green", ls = "--", label = "4 * Exp")
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.grid()
```

```python
# Changeons les limites des axes

z = 2 * y
y3 = 2 * z
plt.plot(x, y, c = "blue", ls = "--", label = "Exp")
plt.plot(x, z, c = "red", ls = "--", label = "2 * Exp")
plt.plot(x, y3, c = "green", ls = "--", label = "4 * Exp")
plt.xlim((6, 10))
plt.ylim((20000, 80000))
plt.title("Fonction Exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.grid()
```

# Faire des subplots

# Approche fonctionnelle

```python
# Graphique avec subplot

plt.subplot(2, 2, 1)
plt.plot(x, y, c = "green", ls = "--", label = "Exp")
plt.subplot(2, 2, 2)
plt.plot(x, y, c = "red", ls = "--", label = "Exp")
plt.subplot(2, 2, 3)
plt.plot(x, y, c = "yellow", ls = "--", label = "Exp")
plt.subplot(2, 2, 4)
plt.plot(x, y, c = "blue", ls = "--", label = "Exp")
plt.subplots_adjust(wspace = 0.5)
```

# Approche orienter objet

```python
# subplots

fig, axes = plt.subplots(2, 1)
axes[0].plot(x, y, c = "green")
axes[1].plot(x, y, c = "red")
```

# 5- Sauvegarder le graphique

```python
# Sauvegarde du graphique dans un fichier

plt.subplot(2, 2, 1)
plt.plot(x, y, c = "green", ls = "--", label = "Exp")
plt.subplot(2, 2, 2)
plt.plot(x, y, c = "red", ls = "--", label = "Exp")
plt.subplot(2, 2, 3)
plt.plot(x, y, c = "yellow", ls = "--", label = "Exp")
plt.subplot(2, 2, 4)
plt.plot(x, y, c = "blue", ls = "--", label = "Exp")
plt.subplots_adjust(wspace = 0.5)
plt.savefig("figure.jpg")
```
