```python
import matplotlib.pyplot as plt
```

# Les donnees

```python
labels = ['abonnes', 'non abonnes']
size = [35, 65]
colors = ['#ff9999', '#66b3ff']
```

# Le graphique

```python
plt.pie(size, labels = labels, colors = colors, autopct = '%1.1f%%', startangle = 90)
```

# Ajout du titre

```python
plt.pie(size, labels = labels, colors = colors, autopct = '%1.1f%%', startangle = 90)
plt.title('Pourcentage de personnes qui regardent les videos sur LeCoinStat et qui ne sont pas abonnes')
```

# Affichage graphique

```python
plt.pie(size, labels = labels, colors = colors, autopct = '%1.1f%%', startangle = 90)
plt.title('Pourcentage de personnes qui regardent les videos sur LeCoinStat et qui ne sont pas abonnes')
plt.show()
```
