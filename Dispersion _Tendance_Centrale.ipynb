# 1- Importation des packages

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

# 2- Chargement de donnees

```python
# Chargement du dataset Iris

iris = sns.load_dataset('iris')

# Affichage des premieres lignes du dataset

print(iris.head())
```

# 3- Calcul des tendances centrales univariees

```python
# Moyenne

mean = iris.mean()
print("Moyenne:\n", mean)

# Mediane

median = iris.median()
print("Mediane:\n", median)

# Mode

mode = iris.mode().iloc(0)
print("\nMode:\n", mode)

# Calcul de la variance pour chaque variable

variance = iris.var()
print("\nVariance:\n", variance)

# Calcul des quantiles

quantiles = iris.quantile([0.25, 0.5, 0.75])
print("\nquantiles:\n", quantiles)
```

```python
# Analyse des tendances centrales de la dispersion
# Nous utilisons la fonction describe() pour obtenir un resumer statistique de chaque variable
# Calcul du resumer statistique avec describe()

sumary = iris.describe()

# Affichage d'un tableau bien formater a l'aide de tabulate

print(tabulate(sumary, headers = 'keys', tablefmt = 'fancy_grid', numalign = 'center'))
```

```python
# Affichage d'un tableau bien formater a l'aide de tabulate

print(tabulate(sumary, headers = 'keys', tablefmt = 'pipe', numalign = 'right'))
```

# 4- Dispersion via des boxplots

```python
# Creation des boxplots univarier pour chaque variable numerique

fig, axes = plt.subplots(1, 4, figsize = (16, 6))

for i, column in enumerate(iris.columns[:-1]):
    sns.boxplot(y = column, data = iris, ax = axes[i])
    axes[i].set_title(f"Boxplot de {column}")
    
plt.tight_layout()
plt.show()
```
