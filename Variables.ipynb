# 1- Importations des packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2- Chargement des donnees

```python
# Chargement des donnees

titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

# 3- Graphiques adapter pour les variables qualitatives

# Graphiques adapter pour les variables nominales

# Diagramme en barres

```python
# Graphique en barres pour la frequence de chaque categorie

plt.figure(figsize = (8, 6)) # Definissons la taille du cadre
ax = sns.countplot(x = "species", data = iris)
plt.title("Nombre de fleurs par categorie")

# Ajoutons le nombre de chaque categorie sur les barres

for p in ax.patches:
    ax.annotate(format(p.get_height(), ".0f"), (p.get_x() + p.get_width() / 2., p.get_height()), ha = "center", va = "center", xytext = (0, 10), textcoords = 'offset points')
```

```python
# Calculons la proportion de chaque categorie

prop_df = iris["species"].value_counts(normalize = True).reset_index()
prop_df.columns = ["species", "proportion"]
plt.figure(figsize = (8, 6)) # Definissons la taille du cadre
ax = sns.barplot(x = "species", y = "proportion", data = prop_df)

# Ajoutons les proportions sur les barres

for p in ax.patches:
    ax.annotate(format(p.get_height(), ".2%"), (p.get_x() + p.get_width() / 2., p.get_height()), ha = "center", va = "center", xytext = (0, 10), textcoords = "offset points")
    plt.show()
```

```python
# Graphique en barres pour la proportion de chaque categorie
# Calculons la proportion de chaque categorie

prop_df = iris["species"].value_counts(normalize = True).reset_index()
prop_df.columns = ["species", "proportion"]

# Graphique en barres pour la frequence de chaque categorie

plt.figure(figsize = (8, 6))
ax = sns.countplot(x = "species", data = iris)
plt.title("Nombre de fleurs par categorie")

# Ajouter le nombre et la proportion de chaque categorie sur les barres

for p, prop, count in zip(ax.patches, prop_df["proportion"], iris["species"].value_counts().sort_index()):
    ax.annotate(f'({count}, {prop: .0%})', (p.get_x() + p.get_width() / 2., p.get_height()), ha = "center", va = "center", xytext = (0, 10), textcoords = "offset points")
    
plt.show()
```

# Camenberg, Graphique en secteurs ou pie chart

```python
# Calculons la proportion et le nombre de chaque categorie
# Calculons la proportion de chaque categorie

prop_df = iris["species"].value_counts(normalize = True).reset_index()
prop_df.columns = ["species", "proportion"]

# Graphique en secteurs pour la frequence de chaque categorie

plt.figure(figsize = (8,6))
ax = iris["species"].value_counts().plot(kind = "pie", autopct = lambda x : f'{x : .0f}% ({x / 100 * len(iris) : .0f})')

# Ajoutons le titre

plt.title("Nombre et proportion de fleurs par categorie")
plt.show()
```

# Diagramme en baton

```python
prop_df = tips["size"].value_counts(normalize = True).reset_index()
prop_df.columns = ["size", "proportion"]

# Graphique en barres pour la frequence de chaque categorie

plt.figure(figsize = (8, 6))
ax = sns.countplot(x = "size", data = tips)
plt.title("Nombre et proportion des clients par taille de groupe")

# Ajoutons le nombre et la proportion de chaque categorie sur les barres

for p, prop, count in zip(ax.patches, prop_df["proportion"], tips["size"].value_counts().sort_index()):
    ax.annotate(f'({count}, {prop : .0%})', (p.get_x() + p.get_width() / 2., p.get_height()),
               ha = "center", va = "center", xytext = (0, 10), textcoords = "offset points")
    
plt.show()
```

# Pie chart

```python
# Nous pouvons faire un boxplot seulement si la variable a entre 3 et 7 modalites
# Calculons la proportion de chaque categorie

prop_df = tips["size"].value_counts(normalize = True).reset_index()
prop_df.columns = ["size", "proportion"]

# Graphique en secteurs pour la frequence de chaque categorie

plt.figure(figsize = (8, 6))
ax = tips["size"].value_counts().plot(kind = "pie", autopct = lambda x : f'{x : .0f}% ({x / 100 * len(iris) : .0f})')

# Ajoutons le titre

plt.title("Nombre et proportion de client par taille de groupe")
plt.show()
```

# 4- Graphiques adapter pour les variables continues

# Graphique adapter pour les variables discretes

# Histogramme

```python
# Creons une variable quantitative discrete

age_var = titanic["age"].dropna().astype(int)
```

```python
# Histogramme

sns.histplot(x = age_var, discrete = True)
plt.title("Distribution des ages sur le titanic")
plt.xlabel("Age")
plt.ylabel("Frequence")
plt.show()
```

# Boxplot

```python
# Boxplot

sns.boxplot(x = age_var)
plt.title("Distribution des ages sur le titanic")
plt.xlabel("age")
plt.show()
```

# Graphique adapter pour les variables continues

# Histogramme

```python
# Creons une variable quantitative continue

quant_var = tips["total_bill"]
```

```python
# Histogramme

sns.histplot(x = quant_var, kde = True)
plt.title("Histogramme de la variable le montant de la facture")
plt.xlabel("valeur")
plt.ylabel("Frequence")
plt.show()
```

# Courbe de densite KDE

```python
# KDE plot

sns.kdeplot(x = quant_var)
plt.title("KDE PLOT de la variable le montant total de la facture")
plt.xlabel("valeur")
plt.ylabel("Densite")
plt.show()
```

```python
# Boite a moustache (Box plot)
# Boite a moustache
# Diagramme en boite

sns.boxplot(x = quant_var)
plt.title("Diagramme en boite de la variable le montant total de la facture")
plt.xlabel("valeur")
plt.show()
```
