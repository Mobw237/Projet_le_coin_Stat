# 6- Fusion et jointure de dataframe

```python
# Creation des dataframes exemples

data1 = {
    "ID" : [1, 2, 3, 4],
    "Prenom" : ["Alice", "Bob", "Charlie", "David"]
}

data2 = {
    "ID" : [3, 4, 5, 6],
    "Ville" : ["Paris", "Lyon", "Marseille", "Toulouse"]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Effectuons une jointure interne (inner join)

inner_join = df1.merge(df2, on = "ID", how = "inner")
#inner_Join

# Effectuons une jointure externe (outer join)

outer_join = df1.merge(df2, on = "ID", how = "outer")
#outer_join

# Effectuons une jointure a gauche (left join)

left_join = df1.merge(df2, on ="ID", how = "left")
#left_join

# Effectuons une jointure droite (right join)

right_join = df1.merge(df2, on = "ID", how = "right")
#right_join

# Affichons les resultats

print("Inner join :", inner_join)
print("\nOuter join : ", outer_join)
print("\nLeft join : ", left_join)
print("\nRight join : ", right_join)
```
