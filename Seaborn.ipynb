# 1- Importation du package

```python
# Importation du package

import seaborn as sns
```

```python
# Liste des datasets disponible dans seaborn

liste_dataset = sns.get_dataset_names()
print(liste_dataset)
```

```python
# Chargement du dataset

iris = sns.load_dataset("iris")
iris.head()
```

```python
# Chargement du dataset titanic

titanic = sns.load_dataset("titanic")
titanic.head()
```

# 2- Graphiques simples avec seaborn

```python
# Nuage de points (scatter plot)

sns.scatterplot(x = "sepal_length", y = "petal_width", data = iris)
```

```python
# Ajoutons des categories

sns.scatterplot(x = "sepal_length", y = "petal_width", data = iris, hue = "species")
```

```python
# Faisons changer la taille des graphiques en fontions des valeurs d'une autre variable

sns.scatterplot(x = "sepal_length", y = "petal_width", data = iris, hue = "species", size = "petal_length")
```

```python
# Changeons la palette de couleur

sns.scatterplot(x = "sepal_length", y = "petal_width", data = iris, hue = "species", size = "petal_length", palette = "YlGnBu")
```

# 3- Quelques graphiques a toujours realiser dans l'analyse de donnees

# Analyser toutes les variables rapidement grace a pairplot

```python
# Pairplot basique

sns.pairplot(iris)
```

```python
# Analyse du pairplot suivant une categorie

sns.pairplot(iris, hue = "species", kind = "kde")
```

# Histogramme

```python
# Tracons un histogramme avec la courbe de densite

sns.histplot(iris.petal_length)
```

# Barplot

```python
# Barplot 

sns.barplot(x = "species", y = "petal_length", data = iris)
```

# Boxplot

```python
# Boxplot entre une variable qualitative et une variable quantitative

sns.boxplot(x = "species", y = "petal_length", data = iris)
```

# Heatmap

```python
# Heatmap : example sur la matrice de correlation

titanic.corr()
sns.heatmap(titanic.corr(), cmap = "YlGnBu", annot = True)
```

# 4- Plotly

```python
# Creation d'un graphique interractif

import plotly.express as px

fig = px.scatter(iris,
                x = "sepal_length",
                y = "sepal_width",
                color = "species",
                size = "petal_length",
                title = "Iris dataset : sepal_length vs sepal_width",
                labels = {"sepal_length " : "Longueur sepal (cm)", "sepal_width" : "Largeur sepal (cm)"})

fig.show()
```

```python
# Chargement du dataset flights

flights = sns.load_dataset("flights")

# Creons un graphique interactif

fig = px.line(flights,
             x = "year",
             y = "passengers",
             color = "month",
             title = "Evolution du nombre de passagers par mois (1949-1960)",
             labels = {"year" : "Annee", "passengers" : "Nombre de passagers", "month" : "mois"})

# Affichons le graphique interactif

fig.show()
```

```python
# Autre exemple
# Chargeons le dataset gapminder

#sns.load_dataset("healthexp")

gapminder = px.data.gapminder()

# Creons un garphique interactif

fig = px.scatter(gapminder,
               x = "gdpPercap",
               y = "lifeExp",
               size = "pop",
               color = "continent",
               hover_name = "country",
               log_x = True,
               range_y = [20, 90],
               animation_frame = "year",
               title = "Evolution de l'esperance de vie et du PIB par habitant (1952-2007)",
               labels = {"gdpPercap" : "PIB par habitant (log)", "lifeExp" : "Esperance de vie", "year" : "Annee", "pop" : "Population", "continent" : "continent"}, height = 600,)
 # Affichons l'apparence du graphique

fig.update_traces(marker = dict(line = dict(width = 1, color = "rgba(0, 0, 0, 0.5)")))
fig.update_layout(showlegend = True, legend = dict(orientation = "h", yanchor = "bottom", y = -0.2, xanchor = "center", x = 0.5))

# Affichons le graphique interactif

fig.show()
```
